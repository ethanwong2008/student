## 1. How I Changed This Year

At the beginning of the year, I tended to stay within my comfort zone by working only with my friends and classmates I already knew. While this felt easy and familiar, it wasn’t fair to others and limited my growth as a developer because strong collaboration skills come from diverse teamwork. Over this trimester, I’ve become more willing to branch out and collaborate with new people. I still want to continue improving by engaging with the entire class rather than a small group.
